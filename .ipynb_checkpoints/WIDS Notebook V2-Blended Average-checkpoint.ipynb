```python
def ensure_data_types(df, num_list, binary_list, string_list):
    # Ensure datatypes are correct
    for i in df.columns:
        if i in num_list:
            df[i] = df[i].astype('Float64')
        elif i in binary_list:
            df[i] = df[i].astype('Int64')
        elif i in string_list:
            df[i] = df[i].astype('category')
    return df
```

```python
def ensure_data_types_V2(df, num_list, binary_list, string_list):
    for i in df.columns:
        if i in num_list:
            df[i] = df[i].astype('Float64')
        elif i in binary_list:
            df[i] = df[i].astype('Float64')
        elif i in string_list:
            df[i] = df[i].astype('str')
    return df
```

```python
michelle_submission = pd.read_csv('DSDJ1.MC13_michelle_submission.csv')
ehsan_submission = pd.read_csv('dsdj2_ehsan_submission.csv')

mikiko_submission = pd.read_csv('MMBAZEL_WIDS2020_submission.csv')
mikiko_knn_submission = pd.read_csv('MMBAZEL_WIDS2020_patientskNN_solution.csv')

susan_submission = pd.read_csv('sw_sample_submission.csv')
```

```python
print(michelle_submission.head())
print(ehsan_submission.head())
print(mikiko_submission.head())
print(mikiko_knn_submission.head())
print(susan_submission.head())
```

```python
michelle_submission = michelle_submission.rename(columns={"hospital_death": "michelle_hospital_death"})
ehsan_submission = ehsan_submission.rename(columns={"hospital_death": "ehsan_hospital_death"})
mikiko_submission = mikiko_submission.rename(columns={"hospital_death": "mikiko_hospital_death"})
mikiko_knn_submission = mikiko_knn_submission.rename(columns={"hospital_death": "mikiko_knn_hospital_death"})
susan_submission = susan_submission.rename(columns={"hospital_death": "susan_hospital_death"})
```

```python
t1 = michelle_submission.merge(ehsan_submission, on='encounter_id', how='left')
t2 = mikiko_submission.merge(mikiko_knn_submission, on='encounter_id', how='left')
t3 = t1.merge(t2, on='encounter_id', how='left')
t4 = t3.merge(susan_submission, on='encounter_id', how='left')
```
